### Change data types 

```python
df['height'] = df['height'].astype('float')
df['weight'] = df['weight'].astype('float')
```

### Removing columns 

```python
# Remove a single column by name(column is null therefore its not necessary) 
df = df.drop('marking', axis=1)
```

```python
df.dtypes
```

### Changing a data type that has special characters

```python
#Values should be numerical but becasue of the $ and , pandas is treating it as a value:

# Step 1: Remove the dollar sign
df['value'] = df['value'].replace({r'\$': ''}, regex=True)

# Step 2: Replace periods with commas (thousands separators)
df['value'] = df['value'].replace({r'\.': ','}, regex=True)

# Step 3: Convert the result to a numeric value by removing commas and converting to float
df['value'] = pd.to_numeric(df['value'].replace({',': ''}, regex=True), errors='coerce')
```
